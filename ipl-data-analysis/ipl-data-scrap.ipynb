```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
import matplotlib.pyplot as plt
```

```python
r = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/indian-premier-league-2023-15129')
soup = bs(r.content)
# print(soup.find('table',attrs={}))
print(soup.prettify())
```

```python
#making the matches table ready

table = soup.find('table',attrs={'class':'ds-w-full ds-table ds-table-xs ds-table-auto ds-w-full ds-overflow-scroll ds-scrollbar-hide'})
row = table.find_all('tr')[1:]
columns = [x.get_text() for x in table.find('tr')]
matches=[]
for r in row:
  data = r.find_all('td')
  d={}
  i=0
  for td in data:
    d[columns[i]]=td.get_text()
    i+=1
  matches.append(d)
matches
```

```python
#creating a batting table for each match

def get_batting(url,all_batting_record):
  r1=requests.get(url)
  soup1=bs(r1.content)
  try:
    title = soup1.find_all('span',attrs={'class':'ds-text-title-xs ds-font-bold ds-capitalize'})[0].get_text()+" VS "+soup1.find_all('span',attrs={'class':'ds-text-title-xs ds-font-bold ds-capitalize'})[1].get_text()
  except:
    title = soup1.find_all('span',attrs={'class':'ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate'})[0].get_text()+" VS "+soup1.find_all('span',attrs={'class':'ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate'})[1].get_text()
  batting_record = soup1.find_all('table',attrs={'class':"ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table"})
  columns = [x.get_text() for x in batting_record[0].select('thead tr th')]
  columns[1] = 'dismissed'
  for batting in batting_record:
      for tr in batting.select('tbody tr')[:-4]:
        if tr['class']==[]:
          d={}
          d['title']=title
          l= [x.get_text() for  x in tr.find_all('td')]
          for index,val in enumerate(l):
            d[columns[index]]=val
          all_batting_record.append(d)

res = [soup.select('table tbody td span a')[i]['href'] for i in range(len(soup.select("table tr td span a"))) if i % 2 != 0]
all_batting_record=[]
for link in res:
  complete_link='https://www.espncricinfo.com'+link;
  get_batting(complete_link,all_batting_record)
all_batting_record
```

```python
#creating a bowling table for each match

def get_bowling(url,all_bowling_record):
  r1=requests.get(url)
  soup1=bs(r1.content)
  try:
    title = soup1.find_all('span',attrs={'class':'ds-text-title-xs ds-font-bold ds-capitalize'})[0].get_text()+" VS "+soup1.find_all('span',attrs={'class':'ds-text-title-xs ds-font-bold ds-capitalize'})[1].get_text()
  except:
    title = soup1.find_all('span',attrs={'class':'ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate'})[0].get_text()+" VS "+soup1.find_all('span',attrs={'class':'ds-text-tight-l ds-font-bold ds-text-typo hover:ds-text-typo-primary ds-block ds-truncate'})[1].get_text()
  bowling_record = soup1.find_all('table',attrs={'class':"ds-w-full ds-table ds-table-md ds-table-auto"})
  columns = [x.get_text() for x in bowling_record[0].select('thead tr th')]
  for bowling in bowling_record:
      for tr in bowling.select('tbody tr'):
        if tr['class']==[]:
          d={}
          d['title']=title
          l= [x.get_text() for  x in tr.find_all('td')]
          for index,val in enumerate(l):
            d[columns[index]]=val
          all_bowling_record.append(d)

res = [soup.select('table tbody td span a')[i]['href'] for i in range(len(soup.select("table tr td span a"))) if i % 2 != 0]
all_bowling_record=[]
for link in res:
  complete_link='https://www.espncricinfo.com'+link;
  get_bowling(complete_link,all_bowling_record)
all_bowling_record
```
